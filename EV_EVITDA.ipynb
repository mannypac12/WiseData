## 코스피 PRICE

```python
prc_dir = 'CSVFile/Price/Korea/Stocks/'
KSE_price = csv.price_reader('KOSPI_FROM_1998_close.csv', prc_dir).loc[:'2018-09-30']
```

```python
KSE_rt = FinancialBacktest(KSE_price)
```

```python
KSE_rt.back_date(KSE_EV_EVIDA_rank == 1, 'Q')
```

```python
tier = {}
for i in range(1, 11):

    tier[f"Tier {i}"] = KSE_rt.backtest(KSE_EV_EVIDA_rank == i, 'Q', 'Q') 
```

```python
Perf_data = pd.DataFrame(tier)
```

```python
test = PfAnalysis(Perf_data)
```

```python
pd.DataFrame([test.totalReturn().rename('Total Return'),
              test.mean_return('A-JUN').rename('Arithmetic Mean Return'),
              test.geo_mean_return('A-JUN').rename('Geometric Mean Return'),
              test.median_return('A-JUN').rename('Median Return'),
              test.volatility().rename('Vol.'),
              test.upordownvol().rename('Upside Vol.'),
              test.upordownvol(option=False).rename('Downside Vol.'),                  
              abs(test.maxDrawDown()).rename('MDD')]).applymap('{: ,.2%}'.format)
```

```python
test.freqReturn(freq='A-JUN').applymap('{: ,.2%}'.format)
```
