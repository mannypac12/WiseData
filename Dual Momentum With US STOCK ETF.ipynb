```python
import dataCleanser as dtc
import matplotlib.pyplot as plt
import matplotlib as mpl


%matplotlib inline
```

```python
objOne = dtc.BasicDualMomentum(file="US_ETF_AND_SORTS.xlsx")   
# test_one = objOne.brc_chan_breakout()
test_two = objOne.compositReturn(windows=30, holding_date=90, start=0, sec=10)
test_two_1 = objOne.compositReturn(windows=30, holding_date=90, start=0, sec=9)
test_two_2 = objOne.compositReturn(windows=30, holding_date=90, start=0, sec=5)
```

```python
for sec in objOne.ret.columns:
    print(sec)
```

```python
font = {'family' : 'KopubDotum',
        'size'   : 22}

mpl.rc('font', **font)
plt.rcParams['axes.unicode_minus'] = False

fig, (ax1, ax2, ax3) = plt.subplots(3,1)
fig.set_size_inches((30,30))

for i, test, ax in zip([10, 9, 5], [test_two, test_two_1, test_two_2], [ax1, ax2, ax3]):

    pf=dtc.PfAnalysis(test)
    (pf.cumReturn().sub(1).div(1/100)).plot(ax=ax)

    
    ax.set_title(f"30 days Return, then holding 90 days, and the num of Max Sector is {i}")
```
