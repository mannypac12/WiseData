```python
def year_factor(price, pershare, st_year=2002, ed_year=2019, non_zero=True):

    dt = pd.DataFrame()

    for year in range(st_year, ed_year):

        dt = pd.concat([dt, price.loc[f"{year}-01-01":f"{year}-03-31"].div(pershare.loc[f"{year-2}-12-31"])])
        dt = pd.concat([dt, price.loc[f"{year}-04-01":f"{year}-12-31"].div(pershare.loc[f"{year-1}-12-31"])])

    if non_zero == True:

        dt[dt < 0] = np.nan

        return dt

    elif non_zero == False:

        return dt

```
