## 예시 Net Income

```python
directory='CSVFile/FinancialData/Korea'
file_name='NET_INCOME_KSE.csv'

net_income = csv.financial_reader(file_name, directory)
```

```python
net_income.tail()
```

```python
net_income_rank = net_income.rank(axis=1, pct=True, ascending=False)
```

```python
net_income_rank.head()
```

```python
for i in range(0, 20):
    
    n_1 = (i+1) / 20
    n = i / 20
    net_income_rank[(n < net_income_rank) & (net_income_rank < n_1)] = n_1 * 20 

    
```
