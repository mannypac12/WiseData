```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from FileOpener.CSVReader.CSVReader import CSVReader as csv

% matplotlib inline
```

## 가격

```python
directory = 'CSVFile/Price/Korea/Stocks'
kse_file = 'KOSPI_FROM_1998_close.csv'
kdq_file = 'KOSDAQ_FROM_1998_close.csv'

kse_price = csv.price_reader(kse_file, directory)
kdq_price = csv.price_reader(kdq_file, directory)
```

# Daily Return

```python
kse_rt = kse_price.div(kse_price.shift(1))
kdq_rt = kdq_price.div(kdq_price.shift(1))
```

## One Month Return

```python
start = '1999-01-01'
end = '2018-12-31'

kse_m_rt = kse_rt.loc[start:end].resample('Q').prod()
kdq_m_rt = kdq_rt.loc[start:end].resample('Q').prod()
```

```python
kse_m_rt[kse_m_rt < 1].sub(kse_m_rt[kse_m_rt < 1].mean(axis=0))
```

```python
def comp(datas):
    
    ret = datas.div(datas.shift(1))
    ret_loss = ret[ret<1]
    return ret_loss.sub(ret_loss.mean(axis=0)) < 0
    
#     return data < np.mean(data)

kse_m_univ = kse_m_rt.apply(comp)
kdq_m_univ = kdq_m_rt.apply(comp)
```
